### Data sanitization
Never trust your data. Surveys can be filled in incorrectly, conventions can change, lots of things can happen.

Let's check the names that are just a single letter. These are likely just initials

```python
babynames[babynames["name"].map(len) <= 1]["name"].unique()
```

This seems like the data were just initials. That's probably not right.

How about names with two letters?

```python
babynames[babynames["name"].map(len) == 2]["name"].unique()
```
